```python
import requests

try:
    with open("api.txt", 'r') as file:
        url_api = file.read().strip()
except FileNotFoundError:
    print("Error: api.txt file not found")
    exit()

try:
    response = requests.get(url_api)
    
    if response.status_code == 200:
        data = response.json()
        
        if data.get("status") == "success" and "data" in data and data["data"]:
            matchlist = data['data']
            print("api status:")
            print(data['info'])
            print("=================================\n")
            
            for match in matchlist:
                print('match name:', match['name'])
                print('match status:', match['status'])
                print('venue:', match['venue'])
                print('date:', match['date'])
                
                if 'score' in match and match['score']:
                    print("score details:\n")
                    for score in match['score']:
                        print('inning:', score['inning'])
                        print('runs:', score['r'])
                        print('wickets:', score['w'])
                        print('overs:', score['o'])
                        print(" ")
                
                print("---------------------------------\n")
        else:
            emsg = data.get("reason", "Unknown error")
            print("no match data retrieved: ", emsg)
    
    else:
        print("Error: failed . Status Codeto fetch information:", response.status_code)
        print("Response:", response.text)
        
except requests.exceptions.RequestException as e:
    print("network error:", e)
```
