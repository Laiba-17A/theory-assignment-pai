```python
import numpy as np

data = []
with open("sensor_data.csv", "r") as file:
    for line in file:
        row = line.strip().split(",")
        data.append(row)

arr = np.array(data, dtype=float)
```

```python
arr[arr == -999] = np.nan
arr[(arr < 0) | (arr > 100)] = np.nan
```

```python
s_mean = np.nanmean(arr, axis=0)
h_median = np.nanmedian(arr, axis=1)
count = np.isnan(arr).sum(axis=0)
worst_sensor = np.argmax(count)
```

```python
minv = np.nanmin(arr)
maxv = np.nanmax(arr)
arr_norm = (arr - minv) / (maxv - minv)
```

```python
np.savetxt("sensor_data_normalized.csv", arr_norm, delimiter=",")
print("worst sensor index:", worst_sensor)
```
