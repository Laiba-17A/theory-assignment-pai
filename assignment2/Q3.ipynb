```python
import pandas as pd

df = pd.read_csv("Titanic-Dataset.csv")

report = []
for col in df.columns:
    dtype = df[col].dtype
    mis_val = df[col].isnull().sum()
    per = (mis_val / len(df)) * 100
    report.append(col + ": dtype=" + str(dtype) + ", missing=" + str(mis_val) + " " + str(round(per, 2)) + " ")

with open("inspect_report.txt", "w") as f:
    f.write("\n".join(report))
```

```python
df['Age'] = df['Age'].fillna(df.groupby(['Pclass', 'Sex'])['Age'].transform('median'))

emb_mode = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(emb_mode)

df = df.drop(columns=['Cabin'])
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df['IsAlone'] = (df['FamilySize'] == 0).astype(int)

df['Age'] = df['Age'].astype('int64')
```

```python
df.to_csv("titanic_cleaned.csv", index=False)

print("data cleaning completed , saved to file")
```
