```python
import pandas as pd

df = pd.read_csv("titanic_cleaned.csv")
tf = pd.read_csv("ticket_fares.csv")

df = df.merge(tf, on="Ticket", how="left")
```

```python
bins = [0, 12, 19, 59, 120]
labels = ['Child', 'Teen', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)
```

```python
surv_sa = df.groupby(['Sex', 'AgeGroup'])['Survived'].mean()
print(surv_sa)
```

```python
with open("report.txt", "w") as f:
    f.write("Hypothesis 1: Women and children first.\n\n")
    f.write("The data shows that women had significantly higher survival rates compared to men across all age groups. Children also demonstrated better survival outcomes than adults. This evidence supports the hypothesis that evacuation protocols prioritized women and children.\n\n")
```

```python
surv_pc = df.groupby('Pclass')['Survived'].mean()
print("Survival by Pclass:")
print(surv_pc)
```

```python
df['FareBin'] = pd.qcut(df['Fare'], q=4, labels=['Low', 'Medium', 'High', 'VeryHigh'])
surv_fb = df.groupby('FareBin')['Survived'].mean()
print("Survival by FareBin:")
print(surv_fb)
```

```python
with open("report.txt", "a") as f:
    f.write("Hypothesis 2: The wealthy had a higher survival rate.\n\n")
    f.write("Method A (Class): Passengers in 1st class showed the highest survival rate, followed by 2nd class, with 3rd class having the lowest rate.\n\n")
    f.write("Method B (Fare): Analysis using fare quartiles reveals that passengers who paid higher fares had notably better survival rates compared to those with lower fares.\n\n")
    f.write("Both methods confirm that wealth positively correlated with survival likelihood, supporting the hypothesis.\n")

print("analysis completed , report saved to file")
```
